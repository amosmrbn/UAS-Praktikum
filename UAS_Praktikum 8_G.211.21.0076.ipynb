```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca file CSV dan membuat dataframe
df = pd.read_csv('stock.csv')

# a. Histogram untuk atribut Purchase Price, Current Price, dan Gain/Loss
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.hist(df['Purchase Price'], bins=10, edgecolor='black')
plt.xlabel('Purchase Price')
plt.title('Distribusi dari Purchase Price')

plt.subplot(1, 3, 2)
plt.hist(df['Current Price'], bins=10, edgecolor='black')
plt.xlabel('Current Price')
plt.title('Distribusi dari Current Price')

plt.subplot(1, 3, 3)
plt.hist(df['Gain/Loss'], bins=10, edgecolor='black')
plt.xlabel('Gain/Loss')
plt.title('Distribusi dari Gain/Loss')

plt.tight_layout()
plt.show()

# b. Atribut Stock Name dan Symbol dengan Dividen / Share tertinggi
div_tertinggi = df.nlargest(5, 'Dividen/Share')[['Stock Name', 'Symbol', 'Dividen/Share']]
print('\n 5 Dividen Tertinggi:')
print(div_tertinggi)

# c. Atribut Stock Name dan Symbol dengan Market Value tertinggi
marketV_tertinggi = df.nlargest(5, 'Market Value')[['Stock Name', 'Symbol', 'Market Value']]
print('\n 5 Market Value Tertinggi:')
print(marketV_tertinggi)

# d. Atribut Stock Name dan Symbol dengan Annual Yield tertinggi
annualY_tertinggi = df.nlargest(7, 'Annual Yield')[['Stock Name', 'Symbol', 'Annual Yield']]
print('\n 7 Annual Yield Tertinggi:')
print(annualY_tertinggi)
```
